```python
import tensorflow as tf
from tensorflow.keras import layers, models
from tensorflow.keras.datasets import fashion_mnist
from tensorflow.keras.utils import to_categorical
import matplotlib.pyplot as plt
from sklearn.metrics import confusion_matrix, classification_report
import seaborn as sns
import numpy as np
```

load fashion

```python
(train_images, train_labels), (test_images, test_labels) = fashion_mnist.load_data()
```

normalize data

```python
train_images = train_images / 255.0
test_images = test_images / 255.0
```

reshaping

```python
train_images = train_images.reshape(60000, 28, 28, 1)
test_images = test_images.reshape(10000, 28, 28, 1)
```

coverting labels to categorical

```python
train_dog = to_categorical(train_labels, 10)
test_dog = to_categorical(test_labels, 10)
```

building CNN model

```python
cnn_model = models.Sequential([

    layers.Conv2D(16, (5,5), activation='relu', input_shape=(28,28,1)),
    layers.MaxPooling2D(pool_size=(2,2)),

    layers.Conv2D(32, (3,3), activation='relu'),
    layers.MaxPooling2D(pool_size=(2,2)),

    layers.Flatten(),

    layers.Dense(64, activation='relu'),
    layers.Dropout(0.4),

    layers.Dense(10, activation='softmax')
])
```

compile model

```python
cnn_model.compile(
    optimizer='adam',
    loss='categorical_crossentropy',
    metrics=['accuracy']
)
```

train model

```python
history = cnn_model.fit(
    train_images,
    train_dog,
    epochs=12,
    batch_size=128,
    validation_split=0.2
)
```

model evaluation

```python
loss_value, accuracy_value = cnn_model.evaluate(test_images, test_dog)

print("Test Accuracy:", accuracy_value)
print("Test Loss:", loss_value)
```

predictions

```python
predictions = cnn_model.predict(test_images)
predicted_classes = np.argmax(predictions, axis=1)
```

confusion matrix

```python
matrix = confusion_matrix(test_labels, predicted_classes)

plt.figure(figsize=(10,8))
sns.heatmap(matrix, annot=True, fmt='d')
plt.title("Confusion Matrix")
plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.show()
```

classification report


```python
print(classification_report(test_labels, predicted_classes))
```

accuracy plot

```python
plt.plot(history.history['accuracy'])
plt.plot(history.history['val_accuracy'])
plt.title("Training vs Validation Accuracy")
plt.xlabel("Epoch")
plt.ylabel("Accuracy")
plt.legend(['Train', 'Validation'])
plt.show()
```

loss plot

```python
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])
plt.title("Training vs Validation Loss")
plt.xlabel("Epoch")
plt.ylabel("Loss")
plt.legend(['Train', 'Validation'])
plt.show()
```

Explanation of Architecture Choices

Two  Convolution Layers were used to help the model learn;
Simple features (edges, lines) and Complex patterns (shapes, clothing textures)

Filters Used;
First layer: 32 filters
Second layer: 64 filters

Kernel Size:
(3×3) is standard and efficient for image feature extraction.
Max Pooling:Reduces image dimensions and computation while preserving important features.

Dropout (0.5):Prevents overfitting by randomly disabling neurons during training.

Softmax Output:Used because this is multi-class classification (10 clothing categories).

3. Batch Normalization - Improves stability and speed.

4. Reduce Dropout Carefully - Too much dropout may hurt learning.
